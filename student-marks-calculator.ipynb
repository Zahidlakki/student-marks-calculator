<!-- Studens marks calculator -->

```python
name = input("enter your name : ")

subjects = int(input("enter number of your subjects : "))
total_marks = 0

for i in range(subjects):
    marks = float(input(f"Enter marks for subject {i+1}: "))
    total_marks += marks
    
percentage = (total_marks/(subjects * 100))*100

if percentage >= 80:
    grade = "A"
elif percentage >= 60:
    grade = "B"
elif percentage >= 40:
    grade = "C"
else:
    grade = "Fail"

print("\n --- Result ---")
print("Student Name : ", name)
print("Total Marks : ", total_marks)
print("percentage : ", percentage)
print("Grade : ", grade)
```
